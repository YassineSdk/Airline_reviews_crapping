```python
import pandas as pd 
import numpy as np 
import os
```

```python
data= {}
for file in os.listdir():
    if file.endswith('.csv'):
        data[file.split('_')[0]]= pd.read_csv(file)
```

```python
import matplotlib.pyplot as plt
```

```python
for key in data.keys():
    if key != 'df':
        print(f'Data name: {key}')
        print(f'data shape{data[key].shape}')
        data[key].isna().sum().plot(kind='bar')
        plt.show()
        print('_'*20)
    
```

```python
ratings_col = [col for col in data['air-arabia'].columns if any('stars' in str(val) for val in data['air-arabia'][col].unique())]
```

```python
ratings_col
```

```python
keys = ['air-arabia', 'air-france', 'royal-air-maroc']
```

```python
for key in keys :
    for col in ratings_col :
        data[key][col] = data[key][col].str.extract(r'([1-9])')
```

```python
data['air-france'].head(3)
```

```python
for key in keys :
    for col in ratings_col :
        mode_value = data[key][col].mode()[0] 
        data[key][col].fillna(mode_value, inplace=True)
```

```python
for key in keys:
        print(f'Data name: {key}')
        print(f'data shape{data[key].shape}')
        data[key].isna().sum().plot(kind='bar')
        plt.show()
        print('_'*20)
```

```python
for key in keys :
    data[key]['text_content'] = data[key]['text_content'].str.replace(r'[^A-Za-z\s]', '', regex=True).str.strip()
```

```python
data.keys()
```

```python
ar_data = data['air-arabia'].copy()
ma_data = data['royal-air-maroc'].copy()
fr_data = data['air-france'] .copy()
```

```python
ar_data
```

```python
for name in ['ar_data','ma_data','fr_data']:
    data = globals()[name]
    data['year'] = data['Date Flown'].str.split(' ').str[1]
    data['year'] = pd.to_datetime(data['year'],format='%Y').dt.strftime('%Y')
    data['Month'] = data['Date Flown'].str.split(' ').str[0]
    
```
